## ESTO ES PARA EL TRATAMIENTO DE LA BBDD DE FARMACIAS PRIVADAS

```python
df_farmacias = pd.read_csv('Maestro FI Marzo 2021.csv',sep=';', encoding='iso-8859-1')
df_farmacias = df_farmacias[df_farmacias['Rut Solicitante'].notna()]
df_farmacias = df_farmacias[df_farmacias['Rut DM'].notna()]
df_farmacias = df_farmacias[(df_farmacias['Ejecutivo']!='CAMBIO DE RUT SOLICITANTE 77145755')&(df_farmacias['Ejecutivo']!='ELIMINADO')&(df_farmacias['Ejecutivo']!='eliminar')]
df_farmacias = df_farmacias.reset_index()
df_farmacias['direccion_completa'] = df_farmacias['Puesto de Descarga']+' , '+df_farmacias['Comuna']+' , '+df_farmacias['Region']
```

```python
# 1 - conveneint function to delay between geocoding calls
locator = Nominatim(user_agent='cenabast')
geocode = RateLimiter(locator.geocode, min_delay_seconds=1)

# 2- - create location column
df_est_cenabast['location'] = df_est_cenabast['concatenacion'].apply(geocode)

# 3 - create longitude, laatitude and altitude from location column (returns tuple)
df_est_cenabast['point'] = df_est_cenabast['location'].apply(lambda loc: tuple(loc.point) if loc else None)

# 4 - split point column into latitude, longitude and altitude columns
df_est_cenabast[['latitude', 'longitude', 'altitude']] = pd.DataFrame(df_est_cenabast['point'].tolist(), index=df_est_cenabast.index)
```
